```python
import numpy as np
import pandas as pd
```

```python
x=np.arange(1,10,2)
y=np.arange(1,10,2)
```

```python
# plt scatter   s=solid (px) c=color(first letter of color)
plt.scatter(x,y,s=25,c='g')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Graph in 2D")

# we can save the image( right click on image)
# or we can save using code
plt.savefig('test.png')
```

```python
y=x*x
```

```python
# plt plot   # third parameter is color

plt.plot(x,y,'g',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("2D diagram")

# In Real EDA, Styling not matter much, seaborn very good than matplotlib
```

```python
# Creating subplots
# with one canvas or diagram we can create multiple plots
```

```python
plt.subplot(2,2,1)  #(nrows, ncols, position) # 4 plots(as 2rows*2cols) # based on the position subplot will appear
plt.plot(x,y,'b*')
plt.subplot(2,2,3)
plt.plot(x,y,'g')
plt.subplot(2,2,2)
plt.plot(x,y,'b')
plt.subplot(2,2,4)
plt.plot(x,y,'y')
```

```python
#Bar plot
# seaborn bar plots are better than matplotlib.pylot

x=[1,10,20]
y=[20,30,10]
```

```python
plt.bar(x,y,color='g')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('2D plot')
```
